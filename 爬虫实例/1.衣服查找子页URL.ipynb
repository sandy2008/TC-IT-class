```python
# 总共103页
for i in range(103):
    page_num = i + 1
    url_target = url_format.format(page_num)
    driver.get(url_target)
    wait = webdriver.support.ui.WebDriverWait(driver, 30)
    time.sleep(2)
    card_elmts = driver.find_elements_by_class_name("grid-product-card")
    for j, card_elmt in enumerate(card_elmts):
        pro_name_elmt = card_elmt.find_element_by_class_name(
            "grid-product-card-display-name")
        pro_price_elmt = card_elmt.find_element_by_class_name(
            "product-card-price__regular")
        sub_link_elmt = card_elmt.find_element_by_class_name(
            "grid-product-card-detail-link")

        pro_name = pro_name_elmt.text
        pro_price = pro_price_elmt.text
        sub_link = sub_link_elmt.get_attribute('href')

        print(pro_name, pro_price, sub_link)

        df.loc[df.shape[0]] = {
            'url': sub_link,
            'name': pro_name,
            'price': pro_price
        }

            #driver.get(sub_url)

    print('第{}页抓取完毕！'.format(page_num))

#sub_url = link_elmt.get_attribute('href')

df.head()
```
